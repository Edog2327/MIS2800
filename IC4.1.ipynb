# ----SPECIFIC FOR PREP/IC/HW ----

```python
%%sql
-- Drop the table if it exists to fix the schema
DROP TABLE IF EXISTS SuperBowl_Jersey;

-- Create the table with STRICT mode correctly applied
CREATE TABLE SuperBowl_Jersey (
  product_id INTEGER PRIMARY KEY,
  name TEXT NOT NULL,
  price REAL NOT NULL
) STRICT;
```

```python
%%sql
-- Verify the new schema
PRAGMA table_info('SuperBowl_Jersey');
```

```python
%%sql
INSERT INTO SuperBowl_Jersey(product_id, name, price)
VALUES
(1, 'Superbowl Jersey', 120.00)
```

```python
%%sql
SELECT * FROM SuperBowl_Jersey WHERE price >50
```

```python
%%sql
UPDATE SuperBowl_Jersey SET price = 99.99 WHERE name='Superbowl Jersey';
DELETE FROM SuperBowl_Jersey WHERE product_id = 1;

SELECT * FROM SuperBowl_Jersey;
```
